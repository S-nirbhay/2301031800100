```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 10, 5]
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 35, 30, 10, 5]
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

# Create subplots (2 rows x 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Matplotlib Visualizations", fontsize=16)

# Line Plot
axs[0, 0].plot(x, y, marker='o', label='Line')
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("X-axis")
axs[0, 0].set_ylabel("Y-axis")
axs[0, 0].legend()
axs[0, 0].grid(True)

# Bar Chart
axs[0, 1].bar(categories, values, color='orange')
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter Plot
axs[1, 0].scatter(scatter_x, scatter_y, color='green')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X")
axs[1, 0].set_ylabel("Y")

# Pie Chart (needs a separate figure)
plt.subplot(2, 2, 4)
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.axis('equal')  # Equal aspect ratio

plt.tight_layout()
plt.show()
```

**This mini project shows how Matplotlib helps turn raw data into visual insights through clear, customizable graphs. It builds a strong foundation for working on more complex data science and analytics projects.**
